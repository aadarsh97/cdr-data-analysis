```python
#removing outliers from the dataset
```

```python
def data_filter(df,col):
  data=df
  colm = data[col]
  data_mean = colm.mean()
  data_std = colm.std()
  data=data[colm<data_mean +3 * data_std]
  return(data)
```

```python
columns = df.columns.tolist()
columns.remove('Phone Number')
columns.remove('Churn')
columns.remove('latitude')
columns.remove('longitude')
columns.remove('time_stamp')
```

```python
for i in columns:
  df=data_filter(df=df,col=i)
```
