```python
new_df = DataFrame()
def data_filter(df,col):
  data=df
  colm = data[col]
  data_mean = colm.mean()
  data_std = colm.std()
  data=data[colm<data_mean +3 * data_std]
  return(data)
```

```python
y = df.columns.tolist()
y.remove('Phone Number')
y.remove('Churn')
y.remove('latitude')
y.remove('longitude')
y.remove('time_stamp')
```

```python
for i in y:
  df=data_filter(df=df,col=i)
```

```python
from random import randrange
covid_result = []
for i in range(len(df)):
    covid_result.append(randrange(0,2))

df["covid_result"] = covid_result
```

```python
df.head()
```

```python
X = df[['Account Length', 'VMail Message', 'Day Mins',
       'Day Calls', 'Day Charge', 'Eve Mins', 'Eve Calls', 'Eve Charge',
       'Night Mins', 'Night Calls', 'Night Charge', 'Intl Mins', 'Intl Calls',
       'Intl Charge', 'CustServ Calls', 'Churn', 'age']]
Y = df['covid_result']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train = scale(X_train)
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
logreg.score(X_train,Y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = logreg.predict(X_test)
print(classification_report(Y_test,y_pred))
```
